```python
# Define a function to update the DataFrame based on the dictionary
def update_df_with_dict(dataframe, dictionary):
    sentence_id = dictionary['sentence_id']  # Get sentence ID from the dictionary
    
    # Check if the sentence-id exists in the DataFrame and update tags
    dataframe.loc[dataframe['sentence-id'] == sentence_id, 'tag_social'] = dictionary['tag_social']
    dataframe.loc[dataframe['sentence-id'] == sentence_id, 'tag_objectness'] = dictionary['tag_objectness']
    dataframe.loc[dataframe['sentence-id'] == sentence_id, 'tag_numerosity'] = dictionary['tag_numerosity']
    dataframe.loc[dataframe['sentence-id'] == sentence_id, 'tag_geometry'] = dictionary['tag_geometry']
    dataframe.loc[dataframe['sentence-id'] == sentence_id, 'tag_temporal'] = dictionary['tag_temporal']
    dataframe.loc[dataframe['sentence-id'] == sentence_id, 'complexity'] = dictionary['tag_complexity']
```

```python
# Apply the dictionary updates to both DataFrames
for dictionary in dict_list:
    update_df_with_dict(df_continuation, dictionary)
    update_df_with_dict(df_truth_value, dictionary)
```

```python
# Melt function for both dataframes
def melt_dataframe(df):
    # Identify the concept tag columns
    tag_columns = [col for col in df.columns if col.startswith('tag_')]
    
    # Identify the columns to keep (non-tag columns)
    id_vars = [col for col in df.columns if col not in tag_columns]
    
    # Melt the dataframe
    melted_df = pd.melt(
        df,
        id_vars=id_vars,
        value_vars=tag_columns,
        var_name='concept_tag',
        value_name='has_tag'
    )
    
    # Filter for True tags and clean up tag names
    melted_df = melted_df[melted_df['has_tag']].copy()
    melted_df['concept_tag'] = melted_df['concept_tag'].str.replace('tag_', '').str.title()
    
    return melted_df

# Melt both dataframes
df_continuation_melted = melt_dataframe(df_continuation)
df_truth_value_melted = melt_dataframe(df_truth_value)
```

```python
# Finally clean up the truth value data

def clean_truth_value_data(df):
    # Create a base dataframe without the item-specific suffix
    base_df = df.copy()
    base_df['item_number'] = base_df['sentence-name'].str.extract(r'truth_value_(\d+)$')
    base_df['base_sentence_name'] = base_df['sentence-name'].str.replace(r'_truth_value_[12]$', '', regex=True)
    
    # Create a pivot table for paired items
    paired_results = base_df.pivot_table(
        index=['sentence-id', 'interval-length', 'model', 'base_sentence_name', 
               'complexity', 'concept_tag', 'has_tag'],
        columns='item_number',
        values='correct',
        aggfunc='first'
    ).reset_index()
    
    # Rename columns for clarity
    paired_results = paired_results.rename(columns={
        '1': 'item1_correct',
        '2': 'item2_correct'
    })
    
    # Add overall correctness
    paired_results['both_correct'] = (paired_results['item1_correct'] & paired_results['item2_correct'])
    
    # Calculate percentage correct for different metrics
    paired_results['percent_items_correct'] = (
        (paired_results['item1_correct'].astype(int) + 
         paired_results['item2_correct'].astype(int)) / 2 * 100
    )
    paired_results['percent_both_correct'] = paired_results['both_correct'].astype(int) * 100
    
    return paired_results

# Create the cleaned dataframe
df_truth_value_clean = clean_truth_value_data(df_truth_value_melted)
```
